## Calculations for Tech Stock Performance Top 5

```python
# Previous day prices and volumes
df['pre_day_apple_price'] = df.apple_price.shift(1)
df['pre_day_apple_vol'] = df.apple_vol.shift(1)
df['pre_day_microsoft_price'] = df.microsoft_price.shift(1)
df['pre_day_microsoft_vol'] = df.microsoft_vol.shift(1)
df['pre_day_google_price'] = df.google_price.shift(1)
df['pre_day_google_vol'] = df.google_vol.shift(1)
df['pre_day_nvidia_price'] = df.nvidia_price.shift(1)
df['pre_day_nvidia_vol'] = df.nvidia_vol.shift(1)
df['pre_day_meta_price'] = df.meta_price.shift(1)
df['pre_day_meta_vol'] = df.meta_vol.shift(1)
```

```python
# Difference between previous and current day prices and volumes
df['price_diff_apple'] = df['apple_price'] - df['pre_day_apple_price']
df['vol_diff_apple'] = df['apple_vol'] - df['pre_day_apple_vol']
df['price_diff_microsoft'] = df['microsoft_price'] - df['pre_day_microsoft_price']
df['vol_diff_microsoft'] = df['microsoft_vol'] - df['pre_day_microsoft_vol']
df['price_diff_google'] = df['google_price'] - df['pre_day_google_price']
df['vol_diff_google'] = df['google_vol'] - df['pre_day_google_vol']
df['price_diff_nvidia'] = df['nvidia_price'] - df['pre_day_nvidia_price']
df['vol_diff_nvidia'] = df['nvidia_vol'] - df['pre_day_nvidia_vol']
df['price_diff_meta'] = df['meta_price'] - df['pre_day_meta_price']
df['vol_diff_meta'] = df['meta_vol'] - df['pre_day_meta_vol']
```

```python
# Differences in percent
df['percent_price_diff_apple'] = df['apple_price'].pct_change()
df['percent_vol_diff_apple'] = df['apple_vol'].pct_change()
df['percent_price_diff_microsoft'] = df['microsoft_price'].pct_change()
df['percent_vol_diff_microsoft'] = df['microsoft_vol'].pct_change()
df['percent_price_diff_google'] = df['google_price'].pct_change()
df['percent_vol_diff_google'] = df['google_vol'].pct_change()
df['percent_price_diff_nvidia'] = df['nvidia_price'].pct_change()
df['percent_vol_diff_nvidia'] = df['nvidia_vol'].pct_change()
df['percent_price_diff_meta'] = df['meta_price'].pct_change()
df['percent_vol_diff_meta'] = df['meta_vol'].pct_change()
```
